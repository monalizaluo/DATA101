```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = np.linspace(0, 12, 24)
y1 = 3 + 6*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 4 + 7*x/9 + np.random.uniform(0.0, 0.5, len(x))

# plot
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

plt.style.use('_mpl-gallery')

n_radii = 10
n_angles = 20

# Make radii and angles spaces
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)

# Define X, Y, Z for tick labels - You'll need to define what these should be
X = ['-0.1', '1', '1.2']  # Example values, replace with your desired labels
Y = ['-0.5', '2', '2.1']  # Example values, replace with your desired labels
Z = ['0.3', '1.4', '3']  # Example values, replace with your desired labels

ax.set(xticklabels=X, # Using the defined variables for tick labels
       yticklabels=Y,
       zticklabels=Z)

plt.show()
```

```python
import matplotlib.pyplot as plt

# defining labels
activities = ['Workout', 'Sleep', 'Study', 'Relax']

# portion covered by each label
slices = [6, 8, 7, 3]

# color for each label
colors = ['y', 'm', 'c', 'b']

# plotting the pie chart
plt.pie(slices, labels=activities, colors=colors,
        startangle=90, shadow=True, explode=(0, 0, 0.1, 0),
        radius=1.2, autopct='%1.1f%%')

# plotting legend with adjusted position
plt.legend(loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1)) # Adjust legend position

plt.title('Percentage of Hours Allocated in Doing Activities')
# showing the plot
plt.show()
```
